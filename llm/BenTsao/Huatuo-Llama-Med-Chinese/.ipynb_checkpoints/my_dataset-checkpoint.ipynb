```python
import pickle
import json
with open("./data/all_data.pickle","rb") as all_data_pickle:
    all_data=pickle.load(all_data_pickle)
    
with open("./data/candidate_second_fine_tune_bert_20.json") as candidate_20:
    candidate=json.load(candidate_20)
```

```python
with open("./data/sim_data.json") as sim_data_json:
    ground_data=json.load(sim_data_json)
```

```python
print(all_data)
```

```python
mention=[key for key in all_data.keys()]
train_name=mention[:378]
test_name=mention[378:]
```

```python
print(train_name[:5])
print(test_name[:5])
```

```python
with open("./data/train_name.pickle","rb")as train_name_pickle:
    train_name=pickle.load(train_name_pickle)
with open("./data/test_name.pickle","rb")as test_name_pickle:
    test_name=pickle.load(test_name_pickle)
```

```python
print(train_name[:5])
print(test_name[:5])
```

```python
instruction="在以下疾病的标准术语“entity”中，与疾病名称“mention”含义相同的是哪一个？"
```

```python
my_test_data=[]
for name in test_name:
    line={}
    instru=instruction
    instru=instru.replace("entity","，".join(candidate[name]))
    instru=instru.replace("mention",name)
    line["instruction"]=instru
    line["input"]=""
    line["output"]=ground_data[name]  
        
    my_test_data.append(line)
```

```python
with open("./data/my_test_data.json","w",encoding="utf-8") as my_test_data_json:
    for line in my_test_data:
        my_test_data_json.write(json.dumps(line,ensure_ascii=False))
        my_test_data_json.write("\n")
```

```python
my_train_data=[]
for name in train_name:
    line={}
    instru=instruction
    instru=instru.replace("entity","，".join(candidate[name]))
    instru=instru.replace("mention",name)
    line["instruction"]=instru
    line["input"]=""
    line["output"]=ground_data[name]
        
    my_train_data.append(line)
```

```python
with open("./data/my_train_data.json","w",encoding="utf-8") as my_train_data_json:
    for line in my_train_data:
        my_train_data_json.write(json.dumps(line,ensure_ascii=False))
        my_train_data_json.write("\n")
```
